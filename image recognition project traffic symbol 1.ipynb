```python
speedlimit50=os.listdir("/Users/macshru/Desktop/python keerthi/traffic symbol/speed limit 50 signs")
```

```python
speedlimit50
```

```python
limit=5
speed_limit50=[None]*limit
j=0
for i in speedlimit50:
    if(j<limit):
        speed_limit50[j]=imread("/Users/macshru/Desktop/python keerthi/traffic symbol/speed limit 50 signs/"+i)
        j+=1
    else:
        break
```

```python
imshow(speed_limit50[1])
```

```python
speed_limit50_bw=[None]*limit
j=0
for i in speedlimit50:
    if(j<limit):
        speed_limit50_bw[j]=rgb2gray(speed_limit50[j])
        j+=1
    else:
        break
```

```python
imshow(speed_limit50_bw[1])
```

```python
speed_limit50_bw[1].shape
```

```python
speed_limit50_bw[2].shape
```

```python
for j in range(5):
    sl=speed_limit50_bw[j]
    speed_limit50_bw[j]=resize(sl,(256,256))
```

```python
speed_limit50_bw[3].shape
```

```python
imshow(speed_limit50_bw[3])
```

```python
len_of_images=len(speed_limit50_bw)
```

```python
len_of_images
```

```python
image_size_speed_limit50=speed_limit50_bw[1].shape
```

```python
image_size_speed_limit50
```

```python
flatten_size_speed_limit50=image_size_speed_limit50[0]*image_size_speed_limit50[1]
```

```python
flatten_size_speed_limit50
```

```python
for i in range(len_of_images):
    speed_limit50_bw[i]=np.ndarray.flatten(speed_limit50_bw[i]).reshape(flatten_size_speed_limit50,1)
```

```python
speed_limit50_bw=np.dstack(speed_limit50_bw)
speed_limit50_bw.shape
```

# change the axis position 

```python
speed_limit50_bw=np.rollaxis(speed_limit50_bw,axis=2,start=0)
```

```python
speed_limit50_bw.shape
```

```python
speed_limit50_bw=speed_limit50_bw.reshape(len_of_images,flatten_size_speed_limit50)
```

```python
speed_limit50_bw
```

# convert array into dataframe

```python
speed_limit50_data=pd.DataFrame(speed_limit50_bw)
```

```python
speed_limit50_data
```

```python
speed_limit50_data["label"]="speedlimit50"
```

```python
speed_limit50_data
```
